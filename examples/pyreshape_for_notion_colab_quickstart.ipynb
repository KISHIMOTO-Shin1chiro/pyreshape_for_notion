## 2. Google Drive をマウント

```python
from pyreshape_for_notion.core import drive_io
my_drive = drive_io.mount_google_drive()
print(my_drive)
```

## 3. 初回投入 (full_export)

0_raw_export フォルダに置いたエクスポートを、parse → split JSON → Notion MD → pcp 分割 → zip バッチ化まで一括処理します。

```python
from pyreshape_for_notion import gemini

result = gemini.full_export(
    raw_export_dir = my_drive / 'Log_Gemini/v1/0_raw_export',
    base_dir       = my_drive / 'Log_Gemini/v1',
)
result
```

## 4. 増分更新 (incremental_export)

0_raw_export を更新したあと、差分のあった会話だけを再生成します。

```python
result = gemini.incremental_export(
    raw_export_dir = my_drive / 'Log_Gemini/v1/0_raw_export',
    base_dir       = my_drive / 'Log_Gemini/v1',
)
result
```

## 5. Notion へのインポート

`_archives/notion_md_zip_batches/` に生成された zip を、Notion の Import 機能から 1 日 1 ファイルのペースで取り込みます (Markdown インポートは 12 時間あたり約 120 ファイルのレート制限があります)。
